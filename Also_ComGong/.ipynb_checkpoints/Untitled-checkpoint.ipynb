# 키, 몸무게, 머리길이, 옷 사이즈

```python
import numpy as np
import random
```

```python
height = input('키 입력:')
weight = input('몸무게 입력:')
length = input('머리길이 : ')
size = input('옷 사이즈: ')

new = [int(height), int(weight), int(length), int(size)]
new = np.array(new)
print(new)
```

```python
w = []
m = []

for i in range(0,50):
    w.append([random.randint(80,200), random.randint(30, 150), random.randint(5,20), random.randrange(75,120,5), 0])
    m.append([random.randint(80,200), random.randint(30, 150), random.randint(5,20), random.randrange(75,120,5), 1])

print(w)
print(m)
```

```python
d =[]

for i in range(0,50):
    d.append([np.sqrt(pow((new[0]-w[i][0]),2)+pow((new[1]-w[i][1]),2)+pow((new[2]-w[i][2]),2)+pow((new[2]-w[i][2]),2)+pow((new[3]-w[i][3]),2)),w[i][4]])
    d.append([np.sqrt(pow((new[0]-m[i][0]),2)+pow((new[1]-m[i][1]),2)+pow((new[2]-m[i][2]),2)+pow((new[2]-m[i][2]),2)+pow((new[3]-m[i][3]),2)),m[i][4]])
d.sort()
print(d)
```

```python
num = input('k를 입력해주세요')
```

```python
cnt = 0
for i in range(int(num)):
    if(d[i][1] == 0):
            cnt +=1
if(cnt>(int(num)/2)):
    print("woman")
else:
    print("man")
```
